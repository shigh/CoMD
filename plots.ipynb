```python
import numpy as np

from glob import glob
from rp import ResultsParser, results_df

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
```

Single processor
================

Performance Baseline
--------------------

```python
cc_files = glob(r"./performance/forceLJ/baseline/cc/*.yaml")
cc = results_df(cc_files)
cc = cc[(cc.function=='force')].sort('totalatoms')

cc_natoms = np.array(cc.totalatoms)
cc_force = np.array(cc.maxtime) / (cc.nsteps.unique() + 1.0)

fig = plt.figure(figsize=(4.0, 4.25))
fig.suptitle("LJ Force Baseline", fontsize=14)
gs = gridspec.GridSpec(1, 1)
ax1 = plt.subplot(gs[0, 0])

ax1.loglog(cc_natoms, cc_force, 'o-b')
ax1.set_xlim(1e3, 1e7)
ax1.set_ylim(1e-4, 1e3)
ax1.set_xlabel("# of atoms")
ax1.set_ylabel("time (s)")

gs.tight_layout(fig, rect=[0, 0, 1, 0.95])
fig.savefig('figs/baseline_forceLJ.pdf', dpi=200)
```

Scaling
--------------

```python
strong_files = glob(r"./scale/strong-T/*.yaml")
strong = results_df(strong_files)
strong = strong[strong.inittemp==0].sort('totalranks')

weak_files = glob(r"./scale/weak-T/*.yaml")
weak = results_df(weak_files)
weak = weak[weak.inittemp==0].sort('totalranks')

nsteps = [strong.nsteps.unique(), weak.nsteps.unique()]
titles = ["Strong Scaling", "Weak Scaling"]
fignames = ['figs/baseline_strong.pdf', 'figs/baseline_weak.pdf']
for i in range(2):
    fig = plt.figure(figsize=(4.0, 4.25))
    fig.suptitle(titles[i], fontsize=14)
    gs = gridspec.GridSpec(1, 1)
    ax1 = plt.subplot(gs[0, 0])

    fs = ["total", "force", "commHalo"]
    markers = ['o-r', 'o-b', 'o-g']

    for j in range(3):
        r = strong[strong.function==fs[j]]
        ax1.loglog(np.array(r.totalranks), 
                   (np.array(r.maxtime) / (r.nsteps.unique() + 1.0)), 
                   markers[j], label=fs[j])
    ax1.legend(loc='upper right', frameon=False)
    ax1.set_xlabel("# of processors")
    ax1.set_ylabel("time (s)")
    gs.tight_layout(fig, rect=[0, 0, 1, 0.95])
    fig.savefig(fignames[i], dpi=200)
```

Performance model
=================

```python
def model(N, n, r, w, c):
    A = 3*w
    B = 3*w+             \
        3*c+3*c+2*c+1*c+ \
        3*c+             \
        2*c+2*c+         \
        3*c+1*c+         \
        3*c+3*c
    C = 1*r
    D = 3*r
    E = 1*c
    return N*(A+n*B/2.0)+C+D+E

tr = 8./(5596.4e6)
tw = tr
tc = 1./(2300.0e6)

cutoff = 2.5*2.315
volume = ((4.0/3.0)*np.pi*cutoff**3)
atomsperm3 = 4.0 / 3.615**3
n = np.ceil(volume * atomsperm3)

model1 = np.array(model(cc_natoms, n, tr, tw, tc)) / (cc.nsteps.unique() + 1.0)
model2 = np.array(model(cc_natoms, 2.0*n, tr, tw, tc)) / (cc.nsteps.unique() + 1.0)

# Fit
#m, b = np.polyfit(np.log10(cc.totalatoms), 
#                  np.log10(cc.maxtime / (cc.nsteps.unique() + 1.0)), 
#                  1)
#print m

# Plot
fig = plt.figure(figsize=(4.0, 4.25))
fig.suptitle("LJ Force Performance Models", fontsize=14)
gs = gridspec.GridSpec(1, 1)
ax1 = plt.subplot(gs[0, 0])

ax1.loglog(cc_natoms, cc_force, 'o-b', label='baseline')
ax1.loglog(cc_natoms, model1, 'o-r', label='model 1')
ax1.loglog(cc_natoms, model2, 'o-g', label='model 2')
ax1.set_xlim(1e3, 1e7)
ax1.set_ylim(1e-4, 1e3)
ax1.legend(loc='upper left', frameon=False)
ax1.set_xlabel("# of atoms")
ax1.set_ylabel("time (s)")

gs.tight_layout(fig, rect=[0, 0, 1, 0.95])
fig.savefig('figs/perfmodel_forceLJ.pdf', dpi=200)
```

Modifications
=============

Choice of Compiler
------------------

```python
gcc_files = glob(r"./performance/forceLJ/baseline/gcc/*.yaml")
gcc = results_df(gcc_files)
gcc = gcc[(gcc.function=='force')].sort('totalatoms')

gcc_natoms = np.array(gcc.totalatoms)
gcc_force = np.array(gcc.maxtime) / (gcc.nsteps.unique() + 1.0)

# Plot
fig = plt.figure(figsize=(8.0, 4.25))
fig.suptitle("Effect of compiler on LJ force", fontsize=14)
gs = gridspec.GridSpec(1, 2)
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])

ax1.loglog(cc_natoms, cc_force, 'o-b', label='baseline, cc')
ax1.loglog(gcc_natoms, gcc_force, 'o-m', label='baseline, gcc')
ax1.set_xlim(1e3, 1e7)
ax1.legend(loc='upper left', frameon=False)
ax1.set_xlabel("# of atoms")
ax1.set_ylabel("time (s)")

ax2.semilogx(cc_natoms, cc_force / gcc_force, 'o-k', label='baseline, cc')
ax2.set_xlim(1e3, 1e7)
ax2.set_ylim(0, 4)
ax2.set_xlabel("# of atoms")
ax2.set_ylabel("speed-up")

gs.tight_layout(fig, rect=[0, 0, 1, 0.95])
fig.savefig('figs/compiler_forceLJ.pdf', dpi=200)
```

Dereferencing pointers to arrays
--------------------------------

```python
cc_opt_files = glob(r"./performance/forceLJ/optim/cc/*.yaml")
cc_opt = results_df(cc_opt_files)
cc_opt = cc_opt[(cc_opt.function=='force')].sort('totalatoms')

cc_opt_natoms = np.array(cc_opt.totalatoms)
cc_opt_force = np.array(cc_opt.maxtime) / (cc_opt.nsteps.unique() + 1.0)

gcc_opt_files = glob(r"./performance/forceLJ/optim/gcc/*.yaml")
gcc_opt = results_df(gcc_opt_files)
gcc_opt = gcc_opt[(gcc_opt.function=='force')].sort('totalatoms')

gcc_opt_natoms = np.array(gcc_opt.totalatoms)
gcc_opt_force = np.array(gcc_opt.maxtime) / (gcc_opt.nsteps.unique() + 1.0)

# Plot
fig = plt.figure(figsize=(8.0, 4.25))
fig.suptitle("Effect of modifications on LJ force", fontsize=14)
gs = gridspec.GridSpec(1, 2)
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])

ax1.loglog(cc_natoms, cc_force, 'o-b', label='baseline, cc')
ax1.loglog(cc_opt_natoms, cc_opt_force, '-*r', label='modified, cc')

ax1.loglog(gcc_natoms, gcc_force, 'o-m', label='baseline, gcc')
ax1.loglog(gcc_opt_natoms, gcc_opt_force, '-*k', label='modified, gcc')

ax1.set_xlim(1e3, 1e7)
ax1.legend(loc='upper left', frameon=False)
ax1.set_xlabel("# of atoms")
ax1.set_ylabel("time (s)")

ax2.semilogx(cc_natoms, cc_force / cc_opt_force, 'o-b', label='cc')
ax2.semilogx(cc_natoms, np.array(gcc_force) / np.array(gcc_opt_force), 'o-m', label='gcc')
ax2.set_xlim(1e3, 1e7)
ax2.set_ylim(1.0, 1.3)
ax2.legend(loc='right', frameon=False)
ax2.set_xlabel("# of atoms")
ax2.set_ylabel("speed-up")

gs.tight_layout(fig, rect=[0, 0, 1, 0.95])
fig.savefig('figs/modified_forceLJ.pdf', dpi=200)
```
