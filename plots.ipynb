Performance Baseline
--------------------

Performance model
=================

Computation-based performance model
-----------------------------------

```python
def model(N, n, r, w, c, k):
    A = 3*w
    B = 3*w+             \
        3*c+3*c+2*c+1*c+ \
        3*c+             \
        2*c+2*c+         \
        3*c+1*c+         \
        3*c+3*c
    C = 1*r
    D = 3*r
    E = 1*c
    # F is for <rc checks that fail
    F = 3*r + 3*c + 3*c + 2*c # No division
    return N*(A + n*B/2.0 + k*F/2.0)+C+D+E

tr = 8./(5596.4e6)
tw = tr
tc = 1./(2300.0e6)

cutoff = 2.5*2.315
volume = ((4.0/3.0)*np.pi*cutoff**3)
atomsperm3 = 4.0 / 3.615**3
n = np.ceil(volume * atomsperm3)

model0 = np.array(model(cc_natoms, n, tr, tw, tc, 0))# / (cc.nsteps.unique() + 1.0)

# Fit
mcc, bcc = np.polyfit(np.log10(cc.totalatoms), np.log10(cc_force), 1)
m0, b0 = np.polyfit(np.log10(cc.totalatoms), np.log10(model0), 1)
print "Slopes"
print "  cc: %1.2f" % mcc
print "  model0: %1.2f" % m0
print "Intercepts"
print "  cc: %1.2f" % bcc
print "  model0: %1.2f" % b0
print "Difference:"
print "  model0: %1.2f" % (bcc-b0)

# Plot
fig = plt.figure(figsize=(4.0, 4.25))
#fig.suptitle("LJ Force Computation-based Performance Model", fontsize=14)
gs = gridspec.GridSpec(1, 1)
ax1 = plt.subplot(gs[0, 0])

ax1.loglog(cc_natoms, cc_force, 'o-b', label='cc baseline')
ax1.loglog(cc_natoms, model0, 'o-r', label='comp-based model')
ax1.set_xlim(1e3, 1e7)
ax1.set_ylim(1e-4, 1e3)
ax1.legend(loc='upper left', frameon=False, fontsize=10)
ax1.set_xlabel("# of atoms")
ax1.set_ylabel("time (s)")

gs.tight_layout(fig, rect=[0, 0, 1, 0.95])
fig.savefig('figs/perfmodel_forceLJ.pdf', dpi=200)
```

Implementation-based performance model
--------------------------------------

```python
k = 375

model1 = np.array(model(cc_natoms, n, tr, tw, tc, k))# / (cc.nsteps.unique() + 1.0)
model2 = np.array(model(cc_natoms, 2.0*n, tr, tw, tc, 2.0*k))# / (cc.nsteps.unique() + 1.0)

# Fit
m1, b1 = np.polyfit(np.log10(cc.totalatoms), np.log10(model1), 1)
m2, b2 = np.polyfit(np.log10(cc.totalatoms), np.log10(model2), 1)
print "Slopes"
print "  model1: %1.2f" % m1
print "  model2: %1.2f" % m2
print "Intercepts"
print "  model1: %1.2f" % b1
print "  model2: %1.2f" % b2
print "Difference:"
print "  model1: %1.2f" % (bcc-b1)
print "  model2: %1.2f" % (bcc-b2)

# Plot
fig = plt.figure(figsize=(4.0, 4.25))
gs = gridspec.GridSpec(1, 1)
ax1 = plt.subplot(gs[0, 0])

ax1.loglog(cc_natoms, cc_force, 'o-b', label='cc baseline')
ax1.loglog(cc_natoms, model0, 'o-r', label='comp-based model')
ax1.loglog(cc_natoms, model1, 'o-y', label='impl-based model')
ax1.loglog(cc_natoms, model2, 'o-g', label='impl-based model, x2')
ax1.set_xlim(1e3, 1e7)
ax1.set_ylim(1e-4, 1e3)
ax1.legend(loc='upper left', fontsize=10, frameon=False)
ax1.set_xlabel("# of atoms")
ax1.set_ylabel("time (s)")

gs.tight_layout(fig, rect=[0, 0, 1, 0.95])
fig.savefig('figs/implmodel_forceLJ.pdf', dpi=200)
```

Modifications
=============

Choice of Compiler
------------------

```python
gcc_files = glob(r"./performance/forceLJ/baseline/gcc/*.yaml")
gcc = results_df(gcc_files)
gcc = gcc[(gcc.function=='force')].sort('totalatoms')

gcc_natoms = np.array(gcc.totalatoms)
gcc_force = np.array(gcc.maxtime) / (gcc.nsteps.unique() + 1.0)

# Fit
mgcc, bgcc = np.polyfit(np.log10(cc.totalatoms), np.log10(gcc_force), 1)
print "Slopes"
print "  gcc: %1.2f" % mgcc
print "Intercepts"
print "  gcc: %1.2f" % bgcc
print "Difference:"
print "  cc: %1.2f" % (bgcc-bcc)
print "  model0: %1.2f" % (bgcc-b0)
print "  model1: %1.2f" % (bgcc-b1)
print "  model2: %1.2f" % (bgcc-b2)

# Plot
fig = plt.figure(figsize=(8.0, 4.25))
gs = gridspec.GridSpec(1, 2)
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])

ax1.loglog(cc_natoms, gcc_force, 'o-m', label='gcc baseline')
ax1.loglog(cc_natoms, cc_force, 'o-b', label='cc baseline')
ax1.loglog(cc_natoms, model0, 'o-r', label='comp-based model')
ax1.loglog(cc_natoms, model1, 'o-y', label='impl-based model')
ax1.loglog(cc_natoms, model2, 'o-g', label='impl-based model, x2')
ax1.set_xlim(1e3, 1e7)
ax1.set_ylim(1e-4, 1e3)
ax1.legend(loc='upper left', fontsize=8, frameon=False)
ax1.set_xlabel("# of atoms")
ax1.set_ylabel("time (s)")

ax2.semilogx(cc_natoms, cc_force / gcc_force, 'o-k', label='cc / gcc')
ax2.set_xlim(1e3, 1e7)
ax2.set_ylim(0, 4)
ax2.legend(loc='lower left', fontsize=8, frameon=False)
ax2.set_xlabel("# of atoms")
ax2.set_ylabel("speed-up")

gs.tight_layout(fig, rect=[0, 0, 1, 0.95])
fig.savefig('figs/compiler_forceLJ.pdf', dpi=200)
```

Dereferencing pointers to arrays
--------------------------------

```python
cc_opt_files = glob(r"./performance/forceLJ/optim/cc/*.yaml")
cc_opt = results_df(cc_opt_files)
cc_opt = cc_opt[(cc_opt.function=='force')].sort('totalatoms')

cc_opt_natoms = np.array(cc_opt.totalatoms)
cc_opt_force = np.array(cc_opt.maxtime) / (cc_opt.nsteps.unique() + 1.0)

gcc_opt_files = glob(r"./performance/forceLJ/optim/gcc/*.yaml")
gcc_opt = results_df(gcc_opt_files)
gcc_opt = gcc_opt[(gcc_opt.function=='force')].sort('totalatoms')

gcc_opt_natoms = np.array(gcc_opt.totalatoms)
gcc_opt_force = np.array(gcc_opt.maxtime) / (gcc_opt.nsteps.unique() + 1.0)

cc_opt_speedup = np.array(cc_force) / np.array(cc_opt_force)
gcc_opt_speedup = np.array(gcc_force) / np.array(gcc_opt_force)

print "Modification speed-up:"
print "  cc: %1.2f - %1.2f" % (np.min(cc_opt_speedup), np.max(cc_opt_speedup))
print "  gcc: %1.2f - %1.2f" % (np.min(gcc_opt_speedup), np.max(gcc_opt_speedup))

# Plot
fig = plt.figure(figsize=(8.0, 4.25))
#fig.suptitle("Effect of modifications on LJ force", fontsize=14)
gs = gridspec.GridSpec(1, 2)
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])

ax1.loglog(cc_natoms, cc_force, 'o-b', label='baseline, cc')
ax1.loglog(cc_opt_natoms, cc_opt_force, '-*r', label='modified, cc')

ax1.loglog(gcc_natoms, gcc_force, 'o-m', label='baseline, gcc')
ax1.loglog(gcc_opt_natoms, gcc_opt_force, '-*k', label='modified, gcc')

ax1.set_xlim(1e3, 1e7)
ax1.legend(loc='upper left', frameon=False, fontsize=10)
ax1.set_xlabel("# of atoms")
ax1.set_ylabel("time (s)")

ax2.semilogx(cc_natoms, cc_opt_speedup, 'o-b', label='cc')
ax2.semilogx(cc_natoms, gcc_opt_speedup, 'o-m', label='gcc')
ax2.set_xlim(1e3, 1e7)
ax2.set_ylim(1.0, 1.3)
ax2.legend(loc='right', frameon=False, fontsize=10)
ax2.set_xlabel("# of atoms")
ax2.set_ylabel("speed-up")

gs.tight_layout(fig, rect=[0, 0, 1, 0.95])
fig.savefig('figs/modified_forceLJ.pdf', dpi=200)
```

What level of speed-up do we need?
---------------------------------

```python
cc_speedup = np.array(cc_opt_force) / np.array(model2)
gcc_speedup = np.array(gcc_opt_force) / np.array(model2)

print "Desired speed-up:"
print "  cc: %1.2f - %1.2f" % (np.min(cc_speedup), np.max(cc_speedup))
print "  gcc: %1.2f - %1.2f" % (np.min(gcc_speedup), np.max(gcc_speedup))

# Plot
fig = plt.figure(figsize=(8.0, 4.25))
#fig.suptitle("Desired speed-up for LJ force", fontsize=14)
gs = gridspec.GridSpec(1, 2)
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])

ax1.loglog(cc_natoms, cc_opt_force, 'o-b', label='modified, cc')
ax1.loglog(gcc_natoms, gcc_opt_force, 'o-m', label='modified, gcc')
ax1.loglog(cc_natoms, model2, 'o-g', label='impl-based model, x2')
ax1.set_xlim(1e3, 1e7)
ax1.legend(loc='upper left', frameon=False, fontsize=10)
ax1.set_xlabel("# of atoms")
ax1.set_ylabel("time (s)")

ax2.semilogx(cc_natoms, cc_opt_force / model2, 'o-b', label='cc / impl-based model, x2')
ax2.semilogx(cc_natoms, gcc_opt_force / model2, 'o-m', label='gcc / impl-based model, x2')
ax2.legend(loc='right', frameon=False, fontsize=10)
ax2.set_xlabel("# of atoms")
ax2.set_ylabel("desired speed-up")

gs.tight_layout(fig, rect=[0, 0, 1, 0.95])
fig.savefig('figs/futurework_forceLJ.pdf', dpi=200)
```
